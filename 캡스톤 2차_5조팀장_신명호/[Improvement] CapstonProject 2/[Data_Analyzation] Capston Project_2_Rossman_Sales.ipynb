```python
df_train['StateHoliday'].replace({'a': 1, 'b': 2, 'c': 3}, inplace=True)
df_test['StateHoliday'].replace({'a': 1, 'b': 2, 'c': 3}, inplace=True)

# 변환 후 데이터를 확인합니다.
print(df_train['StateHoliday'].unique())
print(df_test['StateHoliday'].unique())
```

```python
# 필요한 라이브러리를 임포트합니다.
from sklearn.ensemble import RandomForestRegressor

# X와 Y 데이터를 정의합니다.
X = df_train.drop(['Date','Sales', 'Customers', 'Open'], axis=1)
Y = df_train['Sales']



# 랜덤 포레스트 모델을 생성하고 학습합니다.
model = RandomForestRegressor()
model.fit(X, Y)

# 특성 중요도를 추출합니다.
feature_importances = model.feature_importances_

# 특성 중요도를 시각화합니다.
plt.figure(figsize=(10, 6))
plt.bar(X.columns, feature_importances)
plt.xlabel('Feature')
plt.ylabel('Feature Importance')
plt.title('Feature Importance')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

data_frames = [df_aa, df_ab, df_ac, df_ad, df_ba, df_bb, df_ca, df_cb, df_cc, df_cd]
labels = ['aa', 'ab', 'ac', 'ad', 'ba', 'bb', 'ca', 'cb', 'cc', 'cd']

for df, label in zip(data_frames, labels):
    if df.empty:
        continue

    total_sales = df['Sales'].sum()
    mean_sales = df['Sales'].mean()
    max_sales = df['Sales'].max()
    min_sales = df['Sales'].min()

    total_sales = int(total_sales)

    # 통계 정보 출력
    print(f"{label} 타입의 상점의 총 매출액: {total_sales} 유로")
    print(f"{label} 타입의 상점의 점포당 매출액 평균: {mean_sales:.0f} 유로")
    print(f"{label} 타입의 상점의 최고 매출액: {max_sales:.0f} 유로")
    print(f"{label} 타입의 상점의 최소 매출액: {min_sales:.0f} 유로")

    # 특성 중요도 시각화
    if 'ST_AS' in df:
        # 특성 중요도 추출
        feature_importances = model.feature_importances_

        # 특성명 추출
        feature_names = X.columns

        # 특성 중요도와 특성명을 묶어 데이터프레임 생성
        importance_df = pd.DataFrame({'Feature': feature_names, 'Normalized Importance': feature_importances})
        importance_df = importance_df.sort_values(by='Normalized Importance', ascending=False)

        # ST_AS 칼럼을 제외
        importance_df = importance_df[importance_df['Feature'] != 'ST_AS']

        # 특성 중요도 값 조정
        importance_df['Normalized Importance'] = importance_df['Normalized Importance'].apply(
            lambda x: round(x, 3) if x < 1 else int(x)
        )

        # 특성 중요도 값 출력
        print(f"{label} 타입의 상점 특성 중요도:")
        for index, row in importance_df.iterrows():
            print(f"{row['Feature']}: {row['Normalized Importance']}")

        plt.figure(figsize=(12, 6))
        plt.barh(importance_df['Feature'], importance_df['Normalized Importance'], color='skyblue')
        plt.xlabel('Normalized Importance')
        plt.ylabel('Feature')
        plt.title(f'StoreType {label} Normalized Feature Importance (XGBoost) without ST_AS')
        plt.gca().invert_yaxis()  # 역순으로 정렬

    plt.tight_layout()
    plt.show()
    print()
```
